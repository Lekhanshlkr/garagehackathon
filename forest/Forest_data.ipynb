```python
import pandas as pd
import numpy as np
import pandas_profiling as pp
import matplotlib.pyplot as plt

df = pd.read_csv("forestfires.csv")

df.describe()
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

#plt.hist(df.area)
month_freq = {}
for month in df.month:
    if month not in month_freq:
        month_freq[month] = 0
    month_freq[month] += 1
    
plt.bar([*month_freq], month_freq.values())
plt.show()
```
